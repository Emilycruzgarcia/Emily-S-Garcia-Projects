# Interpretations 
#### Stacked Bar Plot 1: Analyzes of diabetes cases for each (HighBP, HighChol) combination

## X-Axis: HighBP & HighChol Groups
#### Each bar represents a unique combination of the variables:
(0.0, 0.0): Neither high blood pressure nor high cholesterol.

(0.0, 1.0): Only high cholesterol.

(1.0, 0.0): Only high blood pressure.

(1.0, 1.0): Both high blood pressure and high cholesterol.

## Y-Axis: Proportion
The bars show the proportion of individuals within each group for different diabetes statuses.

## Legend: Diabetes Status
#### The colors represent the diabetes outcomes:
Green (0.0): No diabetes/Prediabetes

Orange (1.0): Diabetes present

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grouping: Get proportion of diabetes cases for each (HighBP, HighChol) combination
grouped_data = diabetes_binary_5050.groupby(["HighBP", "HighChol"])["Diabetes_binary"].value_counts(normalize=True).unstack()

# Plot Stacked Bar Chart

# Define custom colors (Green for No Diabetes & Prediabetes, Orangered for Yes)
custom_colors = ["darkgreen","orangered"] 
grouped_data.plot(kind="bar", stacked=True, figsize=(10, 6), color= custom_colors) 

# Title and Labels
plt.title("High Cholesterol & High Blood Pressure vs. Diabetes Status")
plt.xlabel("HighBP & HighChol Groups")
plt.ylabel("Proportion")
plt.legend(title="Diabetes Status (0 = No & Pre, 1 = Yes)")
plt.xticks(rotation=45) #rotates x-axis on the plot
plt.grid(True, linestyle= '--', linewidth=0.5,color='gray')  # Improve readability

# Show Plot
plt.show()
```

# Interpretations
#### Box Plot: Analyzing Age, High Blood Pressure, and High Cholesterol in Diabetes Cases
## X-Axis: Diabetes Status
#### Each box represents a group based on diabetes status:
0: No diabetes or prediabetes

1: Diabetes present

## Y-Axis: Age
#### The box plots show the distribution of ages within each diabetes group.
Higher median ages indicate a greater likelihood of diabetes.

## Legend: HighBP & HighChol Combination
#### The colors represent different combinations of high blood pressure and high cholesterol:
Green (0_0): No High Blood Pressure, No High Cholesterol

Orangered (0_1): No High Blood Pressure, Has High Cholesterol

Orange (1_0): Has High Blood Pressure, No High Cholesterol

Red (1_1): Has High Blood Pressure, Has High Cholesterol

## Age Meanings
1 = 18–24 years

9 = 60–64 years

13 = 80 years or older





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diabetes_binary_5050 = pd.read_csv("Diabetes_binary_5050.csv")

# Column combine with HighBP and HighChol
diabetes_binary_5050['HighBP_HighChol'] = diabetes_binary_5050['HighBP'].astype(str) + '_' + diabetes_binary_5050['HighChol'].astype(str)
# "astype(str) converts into string"

# Create the box plot
plt.figure(figsize=(12, 6))  

# Update the palette keys to match the actual values in 'HighBP_HighChol'
sns.boxplot(x="Diabetes_binary", y="Age", hue="HighBP_HighChol", data=diabetes_binary_5050, 
            palette={"0.0_0.0": "green", "0.0_1.0": "orangered", "1.0_0.0": "orange", "1.0_1.0": "red"})  

plt.title("Age Distribution by Diabetes Status, HighBP, and HighChol")
plt.xlabel("Diabetes Status (0 = No & Pre, 1 = Yes)")
plt.ylabel("Age")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.legend(title="HighBP_HighChol") #(0_0: No, No; 0_1: No, Yes; 1_0: Yes, No; 1_1: Yes, Yes)
plt.show() 
```

#### Each box represents a group based on diabetes status:

## Y-Axis: BMI
#### The box plots show the distribution of BMI (Body Mass Index) within each diabetes group.

Orangered (0_1): No High Blood Pressure, Has High Cholesterol

Orange (1_0): Has High Blood Pressure, No High Cholesterol

Red (1_1): Has High Blood Pressure, Has High Cholesterol

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
diabetes_binary_5050 = pd.read_csv("Diabetes_binary_5050.csv")

# Column combine with HighBP and HighChol
diabetes_binary_5050['HighBP_HighChol'] = diabetes_binary_5050['HighBP'].astype(str) + '_' + diabetes_binary_5050['HighChol'].astype(str)
# "astype(str) converts into string"

# Create the box plot
plt.figure(figsize=(12, 6))

# Update the palette keys to match the actual values in 'HighBP_HighChol'
sns.boxplot(x="Diabetes_binary", y="BMI", hue="HighBP_HighChol", data=diabetes_binary_5050, 
            palette={"0.0_0.0": "green", "0.0_1.0": "orangered", "1.0_0.0": "orange", "1.0_1.0": "red"})

plt.title("BMI Distribution by Diabetes Status, HighBP, and HighChol")
plt.xlabel("Diabetes Status (0 = No & Pre, 1 = Yes)")
plt.ylabel("BMI")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.legend(title="HighBP_HighChol") #(0_0: No, No; 0_1: No, Yes; 1_0: Yes, No; 1_1: Yes, Yes)
plt.show()
```
