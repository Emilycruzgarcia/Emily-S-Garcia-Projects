# Interpretations
#### Stacked Bar Plot 1: Analyzes of diabetes cases for each (HighBP, HighChol) combination
## X-Axis: HighBP & HighChol Groups
#### Each bar represents a unique combination of the variables:
(0.0, 0.0): Neither high blood pressure nor high cholesterol.

(0.0, 1.0): Only high cholesterol.

(1.0, 0.0): Only high blood pressure.

(1.0, 1.0): Both high blood pressure and high cholesterol.

## Y-Axis: Proportion
The bars show the proportion of individuals within each group for different diabetes statuses.

## Legend: Diabetes Status
#### The colors represent the diabetes outcomes:
Green (0.0): No diabetes

Light Orange (1.0) Prediabetes

Orange (2.0): Diabetes present

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
diabetes012 = pd.read_csv('Diabetes_012.csv')


# Grouping: Get proportion of diabetes cases for each (HighBP, HighChol) combination
diabetes012_grouped_data = diabetes012.groupby(["HighBP", "HighChol"])["Diabetes_012"].value_counts(normalize=True).unstack()

# Plot Stacked Bar Chart

# Define custom colors green for no, light orange for predictable, orange for yes
custom_colors = ["darkgreen","coral","orangered"]
diabetes012_grouped_data.plot(kind="bar", stacked=True, figsize=(10, 6), color= custom_colors)

# Title and Labels
plt.title("High Cholesterol & High Blood Pressure vs. Diabetes Status")
plt.xlabel("HighBP & HighChol Groups")
plt.ylabel("Proportion")
plt.legend(title="Diabetes Status (0.0 = No, 1.0 = Pre, 2.0 = Yes)")
plt.xticks(rotation=45) #rotates x-axis on the plot
plt.grid(True, linestyle= '--', linewidth=0.5,color='gray')  # Improve readability

# Show Plot
plt.show()
```

# Interpretations
#### Box Plot: Analyzing Age, High Blood Pressure, and High Cholesterol in Diabetes Cases
## X-Axis: Diabetes Status
#### Each box represents a group based on diabetes status:
0: No diabetes 

1: Prediabetes 

2: Diabetes present

## Y-Axis: Age
#### The box plots show the distribution of ages within each diabetes group.
Higher median ages indicate a greater likelihood of diabetes.

## Legend: HighBP & HighChol Combination
#### The colors represent different combinations of high blood pressure and high cholesterol:
Green (0_0): No High Blood Pressure, No High Cholesterol

Orangered (0_1): No High Blood Pressure, Has High Cholesterol

Orange (1_0): Has High Blood Pressure, No High Cholesterol

Red (1_1): Has High Blood Pressure, Has High Cholesterol

## Age Meanings
1 = 18–24 years

9 = 60–64 years

13 = 80 years or older

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
diabetes012 = pd.read_csv('Diabetes_012.csv')

# Ensure HighBP and HighChol are treated as integers before converting to strings
diabetes012['HighBP_HighChol'] = diabetes012['HighBP'].astype(int).astype(str) + '_' + diabetes012['HighChol'].astype(int).astype(str)

# Define correct palette mapping
palette = {"0_0": "green", "0_1": "orangered", "1_0": "orange", "1_1": "red"}

# Set seaborn style for better visualization
sns.set_style("whitegrid")

# Create the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x="Diabetes_012", y="Age", hue="HighBP_HighChol", data=diabetes012, palette=palette)

# Formatting
plt.title("Age Distribution by Diabetes Status, HighBP, and HighChol")
plt.xlabel("Diabetes Status")
plt.ylabel("Age")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Modify x-axis ticks to show 0, 1, and 2 as categories
plt.xticks([0, 1, 2], ['No Diabetes', 'Prediabetes', 'Diabetes'])

# Update legend to reflect correct labels
plt.legend(title="HighBP & HighChol Combination", labels=[
    "0_0: No BP, No Chol", 
    "0_1: No BP, Yes Chol", 
    "1_0: Yes BP, No Chol", 
    "1_1: Yes BP, Yes Chol"
])

plt.show()

# Data helped formatted by "Gemini" 
```

# Interpretations
#### Box Plot: Analyzing BMI, High Blood Pressure, and High Cholesterol in Diabetes Cases
## X-Axis: Diabetes Status
#### Each box represents a group based on diabetes status:
0: No diabetes or prediabetes

1: Prediabetes

2: Diabetes

## Y-Axis: BMI
#### The box plots show the distribution of BMI (Body Mass Index) within each diabetes group.
Higher median BMI values indicate a greater likelihood of diabetes.

## Legend: HighBP & HighChol Combination
#### The colors represent different combinations of high blood pressure and high cholesterol:
Green (0_0): No High Blood Pressure, No High Cholesterol

Orangered (0_1): No High Blood Pressure, Has High Cholesterol

Orange (1_0): Has High Blood Pressure, No High Cholesterol

Red (1_1): Has High Blood Pressure, Has High Cholesterol

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
diabetes012 = pd.read_csv('Diabetes_012.csv')

# Ensure HighBP and HighChol are treated as integers before converting to strings
diabetes012['HighBP_HighChol'] = diabetes012['HighBP'].astype(int).astype(str) + '_' + diabetes012['HighChol'].astype(int).astype(str)

# Define correct color mapping for HighBP_HighChol combinations
palette = {"0_0": "green", "0_1": "orangered", "1_0": "orange", "1_1": "red"}

# Set seaborn style for better visualization
sns.set_style("whitegrid")

# Create the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x="Diabetes_012", y="BMI", hue="HighBP_HighChol", data=diabetes012, palette=palette)

# Formatting
plt.title("BMI Distribution by Diabetes Status, HighBP, and HighChol")
plt.xlabel("Diabetes Status (0 = No Diabetes, 1 = Prediabetes, 2 = Diabetes)")
plt.ylabel("BMI")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Update legend to reflect correct labels
plt.legend(title="HighBP & HighChol Combination", labels=[
    "0_0: No BP, No Chol", 
    "0_1: No BP, Yes Chol", 
    "1_0: Yes BP, No Chol", 
    "1_1: Yes BP, Yes Chol"
])

plt.show()

# Data helped formatted by "Gemini" 
```
